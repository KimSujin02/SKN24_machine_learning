### 최근접이웃 분류 모델로 훈련 및 평가 해보기

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
# 데이터 로드 및 정제
# 1. input data, target data를 가지고 DataFrame을 만들어서 describe() 확인
input_data = pd.DataFrame(iris_data.data)
print(input_data.describe())
target_data = pd.DataFrame(iris_data.target)
print(target_data.describe())

# 2. 데이터 전처리
#   (1) 입력-정답 데이터를 구분 (O)
fish_input = iris_data.data
fish_label = iris_data.target
#   (2) 훈련-평가 데이터 분리
train_input, test_input, train_label, test_label = train_test_split(
    fish_input,
    fish_label,
    random_state=0,
    stratify=fish_label
)
#   (3) 정규화
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)

train_scaled = (train_input - mean) / std
test_scaled = (test_input - mean) / std

# 모델 만들기
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_label)
kn.score(test_scaled, test_label)
```
