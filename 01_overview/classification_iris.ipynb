# [한번 해보기] 붓꽃 다중분류

붓꽃은 세포핵을 둘러싼 꽃받침(sepal)과 꽃잎(petal)으로 구성되어 있으며, 이러한 특징을 기반으로 붓꽃의 품종을 식별할 수 있습니다.

- Setosa (세토사): Setosa는 붓꽃 중에서 가장 작은 꽃잎과 꽃받침을 가지고 있습니다.
꽃잎과 꽃받침이 비교적 짧고 뾰족한 모습을 갖고 있으며, 주로 흰색 또는 연한 분홍색을 띠고 있습니다.

- Versicolor (버시컬러): Versicolor는 Setosa보다 크고 긴 꽃잎과 꽃받침을 가지고 있습니다.
꽃잎의 색은 보통 연한 보라색이며, 중간 크기의 붓꽃입니다.

- Virginica (버지니카): Virginica는 붓꽃 중에서 가장 크고 긴 꽃잎과 꽃받침을 가지고 있습니다.
꽃잎의 색은 주로 짙은 보라색이며, 다른 품종들에 비해 상대적으로 더 큰 크기를 갖고 있습니다.

```python
from sklearn.datasets import load_iris

iris_data = load_iris()
iris_data.keys()
```

```python
iris_data.data          # input data (= feature data = X data)
iris_data.target        # target data (= label data = y data) / 숫자로 인코딩된 label 데이터
iris_data.target_names  # target data 이름  / label 데이터 (label명)
print(iris_data.DESCR)  # dataset 설명
iris_data.feature_names # input data 이름 / column명
iris_data.filename      # 원본파일명
iris_data.data_module   # 모듈위치
```

```python
type(iris_data.data), type(iris_data.target)
```

### 최근접이웃 분류 모델로 훈련 및 평가 해보기

##### (데이터 로드 및 정제 - 위에서 함)

##### 1. input data, target data를 가지고 DataFrame 만들어서 describe() 확인

```python
import pandas as pd

iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris_df['label'] = iris_data.target

iris_df.describe()
```

##### 2. 데이터 전처리

```python
# (1) 입력-정답 데이터 분리 (이미 되어있음)
```

```python
# (2) 훈련-평가 데이터 분리
from sklearn.model_selection import train_test_split

train_input, test_input, train_label, test_label = \
    train_test_split(iris_data.data, iris_data.target, test_size=.25, random_state=42)
```

```python
# (3) 정규화
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(train_input)
train_scaled = scaler.transform(train_input)
test_scaled = scaler.transform(test_input)
```

##### 3. KNN분류모델 학습

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_scaled, train_label)
```

##### 4. 학습된 모델 평가

```python
kn.score(test_scaled, test_label)
```
